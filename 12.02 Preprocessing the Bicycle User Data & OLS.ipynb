```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import datetime
```

```python
climate = pd.read_csv("climate_1.csv")
```

```python
climate = climate.fillna(0)
```

```python
climate
```

```python
total = pd.read_excel("users.xlsx")
total
```

```python
lst = total['date']
for i in range(len(lst)):
    lst[i] = str(lst[i])
    lst[i] = lst[i].split(' ')[0]
lst
total['date'] = lst
```

```python
climate['weekend'] = 0
climate

for i in range(len(climate.date)):
    climate.date[i] = datetime.datetime.strptime(climate.date[i], '%Y-%m-%d')
    whatday = climate.date[i].weekday()
    if (whatday == 5) or (whatday == 6):
        climate.weekend[i] = 1
climate
```

```python
lst = climate['date']
for i in range(len(lst)):
    lst[i] = str(lst[i])
    lst[i] = lst[i].split(' ')[0]
lst
climate['date'] = lst
```

```python
climate.head(10)
```

```python
total
```

```python
df = pd.merge(total, climate, how='left', left_on='date', right_on='date')
```

```python
df
```

```python
#강수량과 따릉이 이용자의 상관관계

sns.set(rc={'figure.figsize':(20,15)})
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname = "c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family = font_name)

j = sns.jointplot('rain', 'user', data = df, kind = 'reg')
```

```python
#습도와 따릉이 이용자간의 상관관계
j = sns.jointplot('humid', 'user', data = df, kind = 'reg')
```

```python
df = df.set_index('date')
df
```

```python
lst = df['weekend']
```

로그 스케일링은 - 값이 있으면 안된다. 따라서 min max 로 스케일링 하여, 'normalized_df' 로 저장

```python
normalized_df= (df-df.mean())/df.std()
normalized_df
```

```python
normalized_df['weekend'] = lst
normalized_df
```

로그 스케일링하여 결측값 제거 -> 'df' 로 저장

```python
#로그 스케일링
df = df.applymap(lambda x: np.log(x +1))
df
```

```python
sns.set(rc={'figure.figsize':(15,10)})
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname = "c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family = font_name)

corrMatrix = df.corr()
sns.heatmap(corrMatrix, annot=True)
plt.show()
#로그 변환을 한 값
```

```python
sns.set(rc={'figure.figsize':(15,10)})
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname = "c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family = font_name)

corrMatrix = normalized_df.corr()
sns.heatmap(corrMatrix, annot=True)
plt.show()
#min max scale을 한 값
```
