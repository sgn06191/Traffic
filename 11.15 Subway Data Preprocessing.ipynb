```python
import glob
from datetime import date, timedelta

path = "C:/Users/Hannah Kim/Documents/_실증분석연구회/11월 발제/data/*.csv"
date1 = date(2020,1,1)
for fname in glob.glob(path):
    df=pd.read_csv(fname,header= 0,
                        encoding= 'unicode_escape')
    sum_row = df.sum(axis = 0)
    new_row = {'total':sum_row[3], 'general': sum_row[4], 'kid': sum_row[5], 'teen': sum_row[6], 'old': sum_row[8]}
    traffic = traffic.append(new_row, ignore_index = True)
    
```

```python
traffic
```

```python
datelst = pd.date_range(start='1/1/2020', periods=len(traffic), freq='D')
```

```python
datelst = datelst.drop(pd.Timestamp('2020-06-30'))
```

```python
datelst = datelst.insert(len(datelst),pd.Timestamp('2020-10-31'))
```

```python
traffic['date'] = datelst
```
