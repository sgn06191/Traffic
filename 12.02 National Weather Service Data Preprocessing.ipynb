```python
import numpy as np
import pandas as pd
```

```python
climate = pd.read_csv("climate.csv", encoding = 'euc-kr')
```

```python
climate
```

기온과 습도는 그 날짜의 평균 기온과 습도로 처리, 강수량과 적설량은 추후 총 강수량의로 대체

```python
lst = climate['일시']
for i in range(len(lst)):
    lst[i] = lst[i].split(' ')[0]
lst
climate['일시'] = lst
```

```python
climate
```

```python
df = climate.groupby(climate['일시']).mean()
```

```python
df
```

```python
df = df.reset_index()
df
```

```python
df = df.fillna(0)
```

```python
df.to_csv("climate_1.csv", index = False)
```
